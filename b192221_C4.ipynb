```python
import pandas as pd
```

```python
df=pd.read_csv("MELBOURNE_HOUSE_PRICES_LESS.csv")
```

```python
df.head(3)
```

```python
df.tail(3)
```

```python

print(df.shape)
```

```python
df=df.dropna()
```

```python
print(df.shape)
```

```python
rooms=df["Rooms"]
```

```python
print(rooms)
```

```python
price=df["Price"]
```

```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()
plt.hist(x=price,edgecolor="black",bins=10)

plt.xlabel("Price range")
plt.ylabel("Number of Houses")
plt.title("Histogram of Price")
plt.show()
```

```python
rooms_count=df["Rooms"].value_counts()
```

```python
rooms_count=dict(rooms_count)
```

```python
print(rooms_count)
```

```python
a=rooms_count.keys()
b=rooms_count.values()
```

```python
fig=plt.figure(figsize=(4,4))
plt.pie(x=b,labels=a)
plt.legend()
plt.show()
```

```python
regionname_count=df["Regionname"].value_counts()
```

```python
regionname_count=dict(regionname_count)
```

```python
print(regionname_count)
```

```python
r=list(regionname_count.keys())
rc=list(regionname_count.values())
```

```python
fig=plt.figure(figsize=(8,8))
plt.pie(x=rc,labels=r)
plt.legend()
plt.show()
```

```python
#Barplot
type_count=df["Type"].value_counts()
type_count=dict(type_count)
print(type_count)
```

```python
k=list(type_count.keys())
v=list(type_count.values())
```

```python
fig = plt.figure(figsize=(4,4))
plt.bar(x=k,height=v)
plt.show()
```

```python
suburb_count=df["Suburb"].value_counts()
suburb_count=dict(suburb_count)
print(suburb_count)
```

```python
E=list(suburb_count.keys())
F=list(suburb_count.values())
```

```python
fig = plt.figure(figsize=(10,10))
plt.pie(x=F,labels=E)
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(15,15))
plt.bar(x=F,height=E)
plt.show()
```

```python
method_count=df["Method"].value_counts()
method_count=dict(method_count)
print(method_count)
```

```python
y=list(method_count.keys())
z=list(method_count.values())
```

```python
fig = plt.figure(figsize=(4,4))
plt.bar(x=z,height=y,edgecolor="green")
plt.show()
```

```python
price=df["Price"]
price=list(price)
distance=df["Distance"]
distance=list(distance)
```

```python
fig = plt.figure(figsize=(4,4))
plt.plot(distance,price,linestyle="",marker ="o",
        markerfacecolor="green",
        markeredgecolor="black")
plt.xlabel("Distance")
plt.ylabel("Price")
plt.title("Distance vs Price")
plt.show()
```
